```python
import pandas as pd

df = pd.read_parquet("hf://datasets/iamtarun/python_code_instructions_18k_alpaca/data/train-00000-of-00001-8b6e212f3e1ece96.parquet")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from wordcloud import WordCloud


wordsToRemove = spacy.lang.en.stop_words.STOP_WORDS
wordcloud = WordCloud(min_font_size=8, width=800, height=400, background_color='white',stopwords=wordsToRemove).generate(' '.join(df['instruction']))
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("WordCloud of Words Present in Post's Questions", fontsize=16, fontweight='bold', pad=30)
plt.show()
```
